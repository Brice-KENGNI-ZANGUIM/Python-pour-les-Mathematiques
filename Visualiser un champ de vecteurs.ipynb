<h2> On souhaite visualiser dans $[0 , 2\pi]$ la fonction vectorielle : $$ \vec{F}(x,y) = x\cos(x)\,\hat{x} + y\sin(y)\,\hat{y}$$

```python
n = 26

x_r, y_r = np.linspace( 0 , 2*np.pi , n ), np.linspace( 0, 2*np.pi , n )
x , y = np.meshgrid(  x_r , y_r )

f_x , f_y = x*np.sin(x) , y*np.cos(y)

plt.figure(figsize = ( 6.5, 6.5))
Q = plt.quiver(x , y , f_x , f_y , width = n*2e-4 , headlength = 8 , headwidth = 5 )
plt.quiverkey(Q , .84 , 1.025 , 1 , "=1 Kg.m/s", labelpos = "E")

plt.title("Impulsion flow")

plt.axis("scaled")
plt.xlim(-.5,6.8)
plt.ylim(-.5,6.8)
```

<h1> <strong> Formes d'un champ électrique $$ \vec {E}( \vec r) = \frac{kq}{r^2} \hat r = \frac{kq}{r^3}(x\hat x + y\hat y) $$  $$ |\vec {E(\vec r)}| = \frac{kq}{x^2 + y^2} $$

```python
n = 16

x_r, y_r = np.linspace( -1 , 1 , n ), np.linspace( -1, 1 , n )
x , y = np.meshgrid(  x_r , y_r )
r = np.sqrt( x**2 + y**2 )

kq = 1

E_x , E_y = kq*x/r**3 , kq*y/r**3
E_mag = kq/r**2

plt.figure(figsize = ( 6.5, 6.5))
Q = plt.quiver(x , y , E_x , E_y , width = n*2e-4 , headlength = 11 , headwidth = 6 )
#plt.quiverkey(Q , .84 , 1.025 , 1 )#, "=1 Kg.m/s", labelpos = "E")

plt.title("Impulsion flow             ")
plt.axis("scaled")
#plt.xlim(-.5,6.8)
#plt.ylim(-.5,6.8)
```

<h1> <strong> Exemple d'application : Soit le potentiel $$  V(x,y) = \sin(xy)  $$ Le champ électrique associé s'écrit : $$ \vec{E}(x,y) = - \vec {\nabla} V(x,y)  = - \frac{\partial V(x,y) }{\partial x} \hat x  - \frac{\partial V(x,y) }{\partial y} \hat y  = -( y  \hat x + x \hat y ) \cos (xy) $$ 

```python
def V ( n): 
    x = y = np.linspace(-1.2*np.pi , 1.2*np.pi , n)
    x , y = np.meshgrid(x , y)
    return x , y , np.sin(x*y)

def E ( n ) :
    x = y = np.linspace(-1.2*np.pi , 1.2*np.pi , n)
    x , y = np.meshgrid(x , y)
    return x , y , -y*np.cos(x*y) , -x*np.cos(x*y)
```

<h2>  <strong>  Lignes/surfaces équipotentielles

```python
plt.figure( figsize = (9,6))

plt.contour(  *V(500) , 4)
plt.colorbar()
plt.axis("scaled")
plt.xlabel("x (m)")
plt.ylabel("y (m)")

plt.title("Lignes Equipotentielles")

plt.show()
```

<h2>  <strong>  Lignes de Champ électric

```python
plt.figure(figsize = (9,6))

plt.quiver(*E(30))

plt.title("Lignes de champ Electrique")
plt.xlabel("x (m)")
plt.ylabel("y (m)")

plt.axis("scaled")
plt.show()
```
