<h2> <strong> Modules necessaires

```python
import numpy as np
import matplotlib.pyplot as plt
```

<h1> <strong> Représentation 2D

<h2>  <strong> Représenter un vecteur à une position données $$ \vec{F}_{(0,1)} = (1,2) $$

```python
position_x , position_y = 0,1
vecteur_x , vecteur_y = 1 , 2 
plt.quiver(position_x , position_y , vecteur_x , vecteur_y)
plt.show()
```

<h2>  <strong>  Représenter plusieurs vecteurs simultanément : $$ \vec{F}_{(x,0)} = (0,-1) $$

```python
n = 61
```

```python
position_x , position_y = np.linspace(-10,10,n), np.zeros(n)
vecteur_x , vecteur_y = 0 , -1

plt.quiver(position_x, position_y , vecteur_x ,  vecteur_y, width= n*.5e-4)
```

<h2>  <strong>  Champ de Vecteurs 2D : Meshgrid

```python
n = 17
```

```python
position_x , position_y = np.linspace(-10,10,n), np.linspace(-10,10,n)
mesh_position_x, mesh_position_y = np.meshgrid(position_x , position_y)
vecteur_x , vecteur_y = 0 , -1

plt.quiver(mesh_position_x, mesh_position_y , vecteur_x ,  vecteur_y, width= n*2.e-4)
```

```python
position_x , position_y = np.linspace(-10,10,n), np.linspace(-10,10,n)
mesh_position_x, mesh_position_y = np.meshgrid(position_x , position_y)
vecteur_x , vecteur_y = 1 , -1

plt.quiver(mesh_position_x, mesh_position_y , vecteur_x ,  vecteur_y, width= n*2.e-4)
```

```python
position_x , position_y = np.linspace( -10,10,2*n ), np.linspace( -10,10,2*n )
mesh_position_x, mesh_position_y = np.meshgrid( position_x , position_y )
vecteur_x , vecteur_y = np.sin( mesh_position_x ) , np.cos( mesh_position_y )

plt.quiver( mesh_position_x, mesh_position_y , vecteur_x ,  vecteur_y, width= n*1.5e-4 )
```

<h1> <strong> Représentation 3D

<h2>  <strong>  Un seul vecteur

```python
vecteur_1 = [1,0,0]
vecteur_2 = [0,1,0]
vecteur_3 = [0,0,1]

vecteur_origine = [0,0,0]
```

```python
fig = plt.figure(figsize = (8,6))
ax = plt.axes(projection = '3d')

ax.quiver(*vecteur_origine, *vecteur_1, color= "b")
ax.quiver(*vecteur_origine, *vecteur_2, color="g")
ax.quiver(*vecteur_origine, *vecteur_3, color= "r")


ax.set_xlim(-1,2)
ax.set_ylim(-1,2)
ax.set_zlim(0,2)


ax.set_xlabel("axe X")
ax.set_ylabel("axe Y")
ax.set_zlabel("axe Z")
```

<h2>  <strong>  Changement de l'axe de vue

```python
fig = plt.figure(figsize = (8,6))
ax = plt.axes(projection = '3d')

ax.quiver(*vecteur_origine, *vecteur_1, color= "b")
ax.quiver(*vecteur_origine, *vecteur_2, color="g")
ax.quiver(*vecteur_origine, *vecteur_3, color= "r")

ax.set_xlim(-1,2)
ax.set_ylim(-1,2)
ax.set_zlim(-1,2)

ax.set_xlabel("axe X")
ax.set_ylabel("axe Y")
ax.set_zlabel("axe Z")

ax.view_init( 30 , 15 )
```

<h2>  <strong>  Changement de l'axe de vue

```python
n = 7

position_x, position_y, position_z = np.linspace(-10,10,n), np.linspace(-10,10,n), np.linspace(-10,10,n)
position_x, position_y, position_z = np.meshgrid(position_x, position_y, position_z)
vecteur_x, vecteur_y, vecteur_z = position_x, position_y, position_z

fig = plt.figure(figsize = (8,6))
ax = plt.axes(projection = '3d')

ax.quiver( position_x , position_y , position_z , vecteur_x, vecteur_y, vecteur_z , length=0.07, arrow_length_ratio=0.3)

ax.set_xlim(-12,12)
ax.set_ylim(-12,12)
ax.set_zlim(-12,12)

ax.set_xlabel("axe X")
ax.set_ylabel("axe Y")
ax.set_zlabel("axe Z")

ax.view_init( 20 , 25 )
```
